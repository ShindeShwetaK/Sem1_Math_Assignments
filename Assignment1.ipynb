Plot the following graph with sympy library (a) and b) graphs are separate graphs.)

a) f(x)=2x^3-3x+4 and  g(x)=5x-3 the same graph (with title=“my first graph”)

```python
from sympy import *
from sympy.plotting import plot

# Create symbolic variable 'x'
x=symbols('x')

# Define the function you want to plot
f = 2 * x**3 - 3 * x + 4
g = 5 * x - 3

# Create the plot
p=plot(f,g,  show=False,legend=True)

# Set the title
p.title = 'My First Graph'

# Show the plot
p.show()
```

b)z=3x^2 -y^2

```python
from sympy import *
from sympy.plotting import plot3d

# Create symbolic variable 'x'
x,y=symbols('x y')

# Define the function you want to plot
z = 3*x**2 - y**2

#plot3d
plot3d(z)
```
